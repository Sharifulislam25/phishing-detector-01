```python
import pandas as pd
import numpy as np

def decode_qr_url(path):

    img = cv2.imread(
        path,
        cv2.IMREAD_GRAYSCALE
    )

    if img is None:
        return None

    try:
        for obj in decode(img):

            data = obj.data.decode(
                'utf-8',
                errors='ignore'
            )

            return data

    except:
        return None

    return None


# ---------- BENIGN ----------

benign_files = os.listdir(BENIGN_DIR)

benign_records = []

print("Decoding benign QR codes...")

for i, f in enumerate(benign_files[:20000]):

    path = os.path.join(BENIGN_DIR, f)

    url = decode_qr_url(path)

    if url:

        benign_records.append({
            'image_path': path,
            'url': url,
            'label': 0
        })

    if (i + 1) % 500 == 0:
        print(i + 1)


# ---------- MALICIOUS ----------

malicious_files = os.listdir(MALICIOUS_DIR)

malicious_records = []

print("Decoding malicious QR codes...")

for i, f in enumerate(malicious_files[:20000]):

    path = os.path.join(MALICIOUS_DIR, f)

    url = decode_qr_url(path)

    if url:

        malicious_records.append({
            'image_path': path,
            'url': url,
            'label': 1
        })

    if (i + 1) % 500 == 0:
        print(i + 1)


# ---------- FINAL DATASET ----------

qr_dataset = pd.DataFrame(
    benign_records + malicious_records
)

qr_dataset = qr_dataset.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

print("\nQR Dataset Created ✅")
print(qr_dataset.head())

print("\nTotal Samples:", len(qr_dataset))
```

```python
import torch
import torch.nn as nn

from torch.utils.data import (
    Dataset,
    DataLoader
)

from transformers import (
    DistilBertTokenizer
)

from torchvision import transforms

print("Torch + Transformers ready ✅")
```

```python
tokenizer = DistilBertTokenizer.from_pretrained(
    'distilbert-base-uncased'
)

print("Tokenizer loaded ✅")
```

```python
IMAGE_SIZE = 224

train_tf = transforms.Compose([

    transforms.Resize(
        (IMAGE_SIZE, IMAGE_SIZE)
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        [0.485, 0.456, 0.406],
        [0.229, 0.224, 0.225]
    )
])

print("Transforms ready ✅")
```

```python
from PIL import Image

class QRDataset(Dataset):

    def __init__(
        self,
        df,
        tokenizer,
        transforms=None,
        max_len=64
    ):

        self.df = df.reset_index(drop=True)

        self.tokenizer = tokenizer

        self.transforms = transforms

        self.max_len = max_len

    def __len__(self):

        return len(self.df)

    def __getitem__(self, idx):

        row = self.df.iloc[idx]

        # ---- IMAGE ----

        img = Image.open(
            row['image_path']
        ).convert('RGB')

        if self.transforms:
            img = self.transforms(img)

        # ---- URL ----

        url = str(row['url'])

        enc = self.tokenizer(

            url,

            padding='max_length',

            truncation=True,

            max_length=self.max_len,

            return_tensors='pt'
        )

        # ---- LABEL ----

        label = torch.tensor(
            row['label'],
            dtype=torch.long
        )

        return (

            img,

            enc['input_ids'].squeeze(0),

            enc['attention_mask'].squeeze(0),

            label
        )

print("QRDataset class ready ✅")
```

```python
dataset = QRDataset(

    qr_dataset,

    tokenizer,

    transforms=train_tf
)

print("Dataset size:", len(dataset))
```

```python
loader = DataLoader(

    dataset,

    batch_size=16,

    shuffle=True
)

print("DataLoader ready ✅")
```

```python
batch = next(iter(loader))

images, input_ids, attention_masks, labels = batch

print("Images:", images.shape)

print("Input IDs:", input_ids.shape)

print("Attention Masks:", attention_masks.shape)

print("Labels:", labels.shape)
```

```python
import torch.nn as nn

from torchvision.models import vit_b_16

from transformers import DistilBertModel


class DeepGuardQR(nn.Module):

    def __init__(self):

        super().__init__()

        # ---------------- ViT ----------------

        self.vit = vit_b_16(
            weights='IMAGENET1K_V1'
        )

        self.vit.heads = nn.Identity()

        # ---------------- BERT ----------------

        self.bert = DistilBertModel.from_pretrained(
            'distilbert-base-uncased'
        )

        # ---------------- Fusion ----------------

        self.classifier = nn.Sequential(

            nn.Linear(768 + 768, 256),

            nn.ReLU(),

            nn.Dropout(0.3),

            nn.Linear(256, 2)
        )

    def forward(

        self,

        images,

        input_ids,

        attention_masks
    ):

        # ---------- ViT Features ----------

        img_feat = self.vit(images)

        # ---------- BERT Features ----------

        bert_out = self.bert(

            input_ids=input_ids,

            attention_mask=attention_masks
        )

        url_feat = bert_out.last_hidden_state[:,0,:]

        # ---------- Fusion ----------

        fused = torch.cat(

            [img_feat, url_feat],

            dim=1
        )

        # ---------- Classification ----------

        out = self.classifier(fused)

        return out


print("DeepGuardQR model created ✅")
```

```python
device = torch.device(

    'cuda'
    if torch.cuda.is_available()
    else 'cpu'
)

model = DeepGuardQR().to(device)

print(device)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.AdamW(

    model.parameters(),

    lr=2e-5
)

print("Loss + optimizer ready ✅")
```

```python
EPOCHS = 2

model.train()

for epoch in range(EPOCHS):

    total_loss = 0

    for batch in loader:

        images, input_ids, attention_masks, labels = batch

        images = images.to(device)

        input_ids = input_ids.to(device)

        attention_masks = attention_masks.to(device)

        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(

            images,

            input_ids,

            attention_masks
        )

        loss = criterion(
            outputs,
            labels
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch {epoch+1} Loss: {total_loss:.4f}"
    )
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

model.eval()

all_preds = []
all_labels = []

with torch.no_grad():

    for batch in loader:

        images, input_ids, attention_masks, labels = batch

        images = images.to(device)

        input_ids = input_ids.to(device)

        attention_masks = attention_masks.to(device)

        outputs = model(

            images,

            input_ids,

            attention_masks
        )

        preds = torch.argmax(
            outputs,
            dim=1
        )

        all_preds.extend(
            preds.cpu().numpy()
        )

        all_labels.extend(
            labels.numpy()
        )

print("Predictions complete ✅")
```
